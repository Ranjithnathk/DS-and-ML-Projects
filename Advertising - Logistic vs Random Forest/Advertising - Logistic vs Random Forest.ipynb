#### Considering only the numerical values

```python
X=df[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y=df['Clicked on Ad']
```

### Importing Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
X_train
```

```python
y_train
```

### Applying Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test
```

### Logistic Regression Classifier 

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(y_test,lr_pred))
```

```python
print(accuracy_score(y_test,lr_pred))
```

#### Logistic Regression Classifier Model predicted with an accuracy of 97.66%

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
pred_rf=rf.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred_rf))
```

```python
print(accuracy_score(y_test,pred_rf))
```

#### Random Forest Classifier Model predicted with an accuracy of 95.66%

### Logistic Regression Classifier Model wins the battle by predicting the model with an accuracy of 97.66%
